```python
import pandas as pd
```

```python
df = pd.read_csv('./data/Annual_Surface_Temperature_Change.csv', usecols=[1]+list(range(10, 72)))
df
```

```python
df.info()
```

```python
def remove_first_letter(col_name):
     return col_name[1:] if col_name != 'Country' else col_name
    
new_columns = {col: remove_first_letter(col) for col in df.columns}

df.rename(columns=new_columns, inplace=True)

```

```python
df
```

```python
df = df.dropna()
df
```

```python
df.to_csv('./data/output.csv', index=False)
```
